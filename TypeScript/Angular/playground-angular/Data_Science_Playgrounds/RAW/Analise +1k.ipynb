```python
#valores com 14  caracteres cortando os 14 primeiros
values = cadastro_cancel['CARTEIRINHA']
print('(')
for value in values:
    str_value = str(value)
    if(len(str_value) > 14):
        str_value = str_value[0:14]
    if 'nan' in str_value:
        str_value = ""
    str_value = str_value.rjust(14, '0')
    print("'" + str_value + "',")
print(')')
```

```python
#valores com 14  caracteres cortando os 14 últimos
values = cadastro_cancel['CARTEIRINHA']
print('(')
for value in values:
    str_value = str(value)
    if(len(str_value) > 14):
        str_value = str_value[3:]
    if 'nan' in str_value:
        str_value = ""
    str_value = str_value.rjust(14, '0')
    print("'" + str_value + "',")
print(')')
```

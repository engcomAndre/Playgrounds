```python
import requests as rq
import pandas as pd
```

```python
#carregando dados de usuários do hapvida cadastrados no +1k avanz
users_1k_json = rq.get("http://www.hapvida.com.br/mais1k/api/relatorios/usuarios/viver-bem")
```

```python
users = (users_1k_json.json()['users'])
print(len(users))
print((users[0])['user'])
```

```python
#Eliminando caracteres especiais
def remove_special_chars(string):
    treated = ""
    special_chars = "./-"
    for char in string:
        if char.isalnum():
            treated += char            
    return treated
```

```python
#Lendo um usuário
user_1k = (users[0])['user']
nome = (user_1k['Nome']).upper()
dt_nasc = user_1k['Data de nascimento']
cpf = user_1k['CPF']
email = user_1k['Email']
cod_cart = remove_special_chars(user_1k['Código da Carteirinha'])
print("{}, {}, {},{},{}".format(nome,dt_nasc,cpf,email,cod_cart))
```

```python
#Contruindo listas com dados dos usuários
def appends_user_data(nome,dt_nasc,_cpf,email,cod_cart):
    nomes.append(nome)
    dt_nascs.append(dt_nasc)
    cpfs.append(_cpf)
    emails.append(email)
    cod_carts.append(cod_cart)
```

```python
def read_1k_user(user_1k):
    #user_1k = (users[0])['user']
    nome = user_1k['Nome']
    dt_nasc = user_1k['Data de nascimento']
    _cpf = remove_special_chars(str(user_1k['CPF']))
    email = user_1k['Email']
    cod_cart = remove_special_chars(user_1k['Código da Carteirinha'])
    appends_user_data(nome,dt_nasc,_cpf,email,cod_cart)
```

```python
#Lendo todos os usuários fornecidos no serviço
nomes = []
dt_nascs = []
cpfs = []
emails = []
cod_carts = []
for user_1k in users:   
    read_1k_user(user_1k['user'])   
    
```

```python
df = pd.DataFrame({"nomes" : nomes,"dt_nascs":dt_nascs,"cpfs":cpfs,"emails":emails,"cod_carts":cod_carts})
df
```
